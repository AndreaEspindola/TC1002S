# A4 Kmeans

Do Kmeans clustering assuming a number of clusters accorging to your scatter plot

```python
# Import sklearn KMeans
from sklearn.cluster import KMeans

# Define number of clusters
km = KMeans(n_clusters=2)

# Do K-means clustering (assing each point in the dataset to a cluster)
yp = km.fit_predict(dataset[['x1','x2']])


# Print estimated cluster of each point in the dataset
yp
```

Add to your dataset a column with the assihned cluster to each data point

```python
dataset['Cluster1'] = yp
```

Print the number associated to each cluster

```python
dataset.Cluster1.unique()
```

Print the centroids

```python
# Center of each cluster
km.cluster_centers_
```

Print the intertia metric

```python
# Sum of squared error
km.inertia_
```

```python
dataset
```

Plot a scatter plot of your data assigned to each cluster. Also plot the centroids

```python
df0 = dataset[dataset.Cluster1==0]
df1 = dataset[dataset.Cluster1==1]

plt.scatter(df0.x1, df0.x2, label='Cluster 1')
plt.scatter(df1.x1, df1.x2, label='Cluster 2')

plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], color='magenta', marker='*', label='Centroides', s=256)

plt.title('x1 vs x2')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.show()
```

# A4 Elbow plot

Compute the Elbow plot

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit_predict(dataset[['x1','x2']])
    sse.append(km.inertia_)
```

```python
plt.plot(k_rng,sse)

plt.title('Elbow plot')
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.show()
```
